```python
driver.quit()
```

```python
weeks = driver.find_elements_by_css_selector("div.col")
```

```python
#-*- coding: utf-8 -*-
for week in weeks:
    titles = week.find_elements_by_css_selector("a.title")
    titles = [title.text for title in titles]
    for t in titles:
        print t
```
